```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
print("Ride Data")
print(ride_df.head())
print("-----------------------------------------------------------------")
print("City Data")
print(city_df.head())
```

```python
# Combine the data into a single dataset
city_ride_df = pd.merge(ride_df, city_df, on="city", how="left")

# Display the data table for preview

city_ride_df.head()
```

```python
# divide the data frame into 3 sub data frames based on city type

urban_data = city_ride_df.loc[city_ride_df["type"] == "Urban", :]

suburban_data = city_ride_df.loc[city_ride_df["type"] == "Suburban", :]

rural_data = city_ride_df.loc[city_ride_df["type"] == "Rural", :]
```

```python
# urban data grouped by city
urban_data_grouped = urban_data.groupby(["city"])

city_type = urban_data.set_index("city")["type"]

average_fare_urban= urban_data_grouped.mean()["fare"]

rides_count_urban = urban_data_grouped.count()["ride_id"]

driver_count_urban = urban_data_grouped.mean()["driver_count"]

citywise_urban_data = pd.DataFrame({"Average Fare ($) Per City": average_fare_urban, 
                           "Total Number of Rides Per City": rides_count_urban,
                            "Total Number of Drivers Per City": driver_count_urban})
```

```python
# suburban data grouped by city
suburban_data_grouped = suburban_data.groupby(["city"])

average_fare_suburban= suburban_data_grouped.mean()["fare"]

rides_count_suburban = suburban_data_grouped.count()["ride_id"]

driver_count_suburban = suburban_data_grouped.mean()["driver_count"]

citywise_suburban_data = pd.DataFrame({"Average Fare ($) Per City": average_fare_suburban, 
                           "Total Number of Rides Per City": rides_count_suburban,
                            "Total Number of Drivers Per City": driver_count_suburban})
```

```python
# rural data grouped by city
rural_data_grouped = rural_data.groupby(["city"])

average_fare_rural= rural_data_grouped.mean()["fare"]

rides_count_rural = rural_data_grouped.count()["ride_id"]

driver_count_rural = rural_data_grouped.mean()["driver_count"]

citywise_rural_data = pd.DataFrame({"Average Fare ($) Per City": average_fare_rural, 
                           "Total Number of Rides Per City": rides_count_rural,
                            "Total Number of Drivers Per City": driver_count_rural})
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_urban = citywise_urban_data["Total Number of Rides Per City"]
y_urban = citywise_urban_data["Average Fare ($) Per City"]

x_suburban = citywise_suburban_data["Total Number of Rides Per City"]
y_suburban = citywise_suburban_data["Average Fare ($) Per City"]

x_rural = citywise_rural_data["Total Number of Rides Per City"]
y_rural = citywise_rural_data["Average Fare ($) Per City"]
```

```python
plt.figure(figsize=(6,4))

# Incorporate a text label regarding circle size
size_urban = citywise_urban_data["Total Number of Drivers Per City"]*10
size_suburban = citywise_suburban_data["Total Number of Drivers Per City"]*10
size_rural = citywise_rural_data["Total Number of Drivers Per City"]*10

# Build the scatter plots for each city type
plt.scatter(x_urban, y_urban, marker="o", facecolors="coral", edgecolors="black", s= size_urban, label= "Urban", alpha=0.75)
plt.scatter(x_suburban, y_suburban, marker="o", facecolors="skyblue", edgecolors="black", s= size_suburban, label= "Suburban", alpha=0.75)
plt.scatter(x_rural, y_rural, marker="o", facecolors="gold", edgecolors="black", s= size_rural, label= "Rural", alpha=0.75)

# Incorporate the other graph properties
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)", fontsize=16)
plt.xlabel("Total Number of Rides (Per City)", fontsize=14)
plt.ylabel("Average Fare ($)", fontsize=14)
plt.text(42,32, "Note:\n Circle Size correlates with driver count per city", fontsize=14)

# Create a legend
lgnd = plt.legend(loc="best", shadow=True, title="City Types", fontsize=10,  fancybox=True)
title = lgnd.get_title()
title.set_fontsize(12)

# change font size of legends
for handle in lgnd.legendHandles:
    handle.set_sizes([30.0])


# Save Figure
plt.savefig("Images/Ride_Sharing.png", bbox_inches='tight')

# Show Figure
plt.show()
```

## Total Fares by City Type

```python
plt.figure(figsize=(6,6))

# Calculate Type Percents
total_fare_urban = urban_data_grouped.sum()["fare"].sum()
total_fare_suburban = suburban_data_grouped.sum()["fare"].sum()
total_fare_rural = rural_data_grouped.sum()["fare"].sum()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [total_fare_urban, total_fare_suburban, total_fare_rural]
colors = ["coral", "xkcd:lightblue", "gold"]
explode = (0.1, 0.0, 0.0)

plt.pie(sizes, autopct="%1.1f%%", explode=explode, shadow=True, labels=labels, 
        colors= colors, counterclock =False, startangle=150, textprops={'fontsize': 14})


plt.title("% of Total Fares by City Type", fontsize = 16)

# Save Figure
plt.savefig("Images/Total_Fare.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
plt.figure(figsize=(6,6))

# Calculate Ride Percents
total_rides_urban = urban_data_grouped.count()["ride_id"].sum()
total_rides_suburban = suburban_data_grouped.count()["ride_id"].sum()
total_rides_rural = rural_data_grouped.count()["ride_id"].sum()

# Build Pie Chart
labels = ["Urban", "Suburban", "Rural"]
sizes = [total_rides_urban, total_rides_suburban, total_rides_rural]
colors = ["coral", "skyblue", "gold"]
explode = (0.1, 0.0, 0.0)

plt.pie(sizes, autopct="%1.1f%%", explode=explode, shadow=True, labels=labels, 
        colors=colors, counterclock =False, startangle=150, textprops={'fontsize': 14})

plt.title("% of Total Rides by City Type", fontsize = 16)

# Save Figure
plt.savefig("Images/Total_Rides.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
plt.figure(figsize=(6,6))

# Calculate Driver Percents
total_drivers_urban = urban_data_grouped.mean()["driver_count"].sum()
total_drivers_suburban = suburban_data_grouped.mean()["driver_count"].sum()
total_drivers_rural = rural_data_grouped.mean()["driver_count"].sum()

# Build Pie Charts

labels = ["Urban", "Suburban", "Rural"]
sizes = [total_drivers_urban, total_drivers_suburban, total_drivers_rural]
colors = ["coral", "skyblue", "gold"]

explode = (0.1, 0.0, 0.0)
plt.pie(sizes, autopct="%1.1f%%", explode=explode, shadow=True, labels=labels, 
        colors=colors, counterclock =False, startangle=150, textprops={'fontsize': 14})

plt.title("% of Total Drivers by City Type", fontsize = 16)

# Save Figure
plt.savefig("Images/Total_Drivers.png")

# Show Figure
plt.show()
```
